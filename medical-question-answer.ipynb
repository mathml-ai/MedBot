```python
from transformers import AutoModelForCausalLM, AutoTokenizer, TrainingArguments, BitsAndBytesConfig
from datasets import load_dataset
import torch
from peft import LoraConfig
from trl import SFTTrainer
from kagglehub import model_download, KaggleDatasetAdapter, load_dataset

# Load and verify dataset
dataset = load_dataset(
    KaggleDatasetAdapter.HUGGING_FACE,
    "pythonafroz/medquad-medical-question-answer-for-ai-research",
    "medquad.csv",
).filter(lambda x: x['answer'] is not None)  # Remove empty answers
```

```python
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_use_double_quant=True,
    bnb_4bit_compute_dtype=torch.bfloat16
)

model_path =  model_download('qwen-lm/qwen-3/transformers/1.7b')

# Load base model
model = AutoModelForCausalLM.from_pretrained(
    model_path,
    quantization_config=bnb_config,
    device_map="auto",
    trust_remote_code=True
)
tokenizer = AutoTokenizer.from_pretrained(model_path)
```

```python
messages = [
    {"role" : "user", "content" : "What is at risk for Diabetic Retinopathy? Answer it concisely so that a normal person who is not related to medical field can also understand it properly."}
]
text = tokenizer.apply_chat_template(
    messages,
    tokenize = False,
    add_generation_prompt = True,
    enable_thinking = True,
)

from transformers import TextStreamer
_ = model.generate(
    **tokenizer(text, return_tensors = "pt").to("cuda"),
    max_new_tokens = 2048,
    temperature = 0.7, top_p = 0.8, top_k = 20,
    streamer = TextStreamer(tokenizer, skip_prompt = True),
)
```

```python
import random

# Take first 5 records from the dataset as few-shot examples
few_shot_examples = [
    {"question": q, "answer": a}
    for q, a in zip(dataset["question"][:5], dataset["answer"][:5])
]

def format_data(examples):
    result = []
    for q, a in zip(examples['question'], examples['answer']):
        
        # Build system instruction
        system_msg = {
            "role": "system",
            "content": (
                "You are an expert assistant. Your goal is to provide clear, detailed, "
                "and well-structured answers to the user's questions. "
                "You will be shown several example Q&A pairs — study them carefully to learn the expected depth, style, and reasoning. "
                "After the examples, answer the final question in the same manner, with accuracy, clarity, and completeness."
            )
        }
        
        # Add the fixed 5 examples
        few_shots_chat = []
        for ex in few_shot_examples:
            few_shots_chat.append({"role": "user", "content": ex["question"]})
            few_shots_chat.append({"role": "assistant", "content": ex["answer"]})
        
        # Append the target Q/A pair
        full_chat = [system_msg] + few_shots_chat + [
            {"role": "user", "content": q},
            {"role": "assistant", "content": a}
        ]
        
        # Apply Qwen's chat template
        result.append(
            tokenizer.apply_chat_template(full_chat, tokenize=False)
        )
    
    return {"text": result}

processed_dataset = dataset.map(format_data, batched=True)
```

```python
question = "What are common symptoms of typhoid?"

# Zero-shot prompt
zero_shot_chat = [
    {"role": "system", "content": "You are a knowledgeable medical assistant. Answer the question accurately and concisely."},
    {"role": "user", "content": question}
]

zero_shot_prompt = tokenizer.apply_chat_template(zero_shot_chat, tokenize=False)
inputs = tokenizer(zero_shot_prompt, return_tensors="pt").to(model.device)

output_ids = model.generate(**inputs, max_new_tokens=200, temperature=0.7)
zero_shot_answer = tokenizer.decode(output_ids[0], skip_special_tokens=True)
print("\nZero-shot answer:\n", zero_shot_answer)
```

```python
# Use the same 5 fixed examples we made earlier
few_shot_examples = [
    {"question": q, "answer": a}
    for q, a in zip(dataset["question"][:5], dataset["answer"][:5])
]

# Build few-shot chat
system_msg = {
    "role": "system",
    "content": (
        "You are a knowledgeable medical assistant. "
        "Below are example Q&A pairs — study them for style and depth, "
        "then answer the final question in a similar detailed manner."
    )
}

few_shots_chat = [system_msg]
for ex in few_shot_examples:
    few_shots_chat.append({"role": "user", "content": ex["question"]})
    few_shots_chat.append({"role": "assistant", "content": ex["answer"]})

# Add the target question
few_shots_chat.append({"role": "user", "content": question})

few_shot_prompt = tokenizer.apply_chat_template(few_shots_chat, tokenize=False)
inputs = tokenizer(few_shot_prompt, return_tensors="pt").to(model.device)

output_ids = model.generate(**inputs, max_new_tokens=200, temperature=0.7)
few_shot_answer = tokenizer.decode(output_ids[0], skip_special_tokens=True)
print("\nFew-shot answer:\n", few_shot_answer)
```

```python
from datasets import Dataset
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig
from peft import prepare_model_for_kbit_training, LoraConfig, get_peft_model,EarlyStoppingCallback
from trl import SFTTrainer
import torch

# Assuming processed_dataset is already a HF Dataset with 'text'
dataset = processed_dataset.shuffle(seed=42)
train_test = dataset.train_test_split(test_size=0.1)
train_dataset = train_test["train"]
eval_dataset = train_test["test"]

# ✅ LoRA Config
lora_config = LoraConfig(
    r=64,
    lora_alpha=64,
    lora_dropout=0.25,
    target_modules=["q_proj", "k_proj", "v_proj", "o_proj", "gate_proj", "up_proj", "down_proj"],
    bias="none",
    task_type="CAUSAL_LM"
)






```

```python
training_args = TrainingArguments(
    output_dir="./qwen-medical",
    per_device_train_batch_size=2,
    gradient_accumulation_steps=4,
    learning_rate=2e-4,
    max_steps=30,
    fp16=True,
    optim="adamw_8bit",
    report_to="none",
    logging_steps=2,
    neftune_noise_alpha=5,
    gradient_checkpointing=True
)
```

```python
early_stopping_callback = EarlyStoppingCallback(
    early_stopping_patience=5,  # Stop if eval_loss doesn't improve for 3 evaluations
    early_stopping_threshold=0.01, # Consider improvement only if loss decreases by at least 0.01
)
```

```python
trainer = SFTTrainer(
    model=model,
    processing_class=tokenizer,
    train_dataset=train_dataset,
    eval_dataset=eval_dataset,
    formatting_func=lambda x: x["text"],
    
    args=training_args,
    peft_config=lora_config,
    callbacks=[early_stopping_callback]
)

trainer.train()
```
